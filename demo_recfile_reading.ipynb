```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import mxnet as mx 
```

```python
def decodeBox(yolobox, size, dscale):
    i, j, x, y, w, h = yolobox
    cxt = j*dscale + x*dscale
    cyt = i*dscale + y*dscale
    wt = w*size
    ht = h*size
    # true location in image
    return [cxt, cyt, wt, ht]

def bboxdraw(img, label, dscale=32):
    assert label.shape == (7,7,9)
    size = img.shape[1]
    ilist, jlist = np.where(label[:,:,0]==1)
    
    # Create figure and axes
    fig,ax = plt.subplots(1)
    ax.imshow(img)
    for i,j in zip(ilist, jlist): 
        cx,cy,w,h = label[i,j,1:5]
        cxt, cyt, wt ,ht = decodeBox([i, j, cx,cy,w,h], size, dscale)
        print(cxt, cyt, wt ,ht)
        # Create a Rectangle patch
        rect = patches.Rectangle((cxt-wt/2,cyt-ht/2), wt,ht,linewidth=1,edgecolor='r',facecolor='none')

        # Add the patch to the Axes
        ax.add_patch(rect)

    plt.show()
```

```python
data = mx.io.ImageRecordIter(path_imgrec='DATA_rec/drive_small.rec',
                             data_shape=(3,224,224),
                             label_width=7*7*9, 
                             batch_size=2,)
```

```python
num =5 # the number of batch
img_num = 0
data.reset()
for i in range(num):
    batch = data.next()
    for j in range(len(batch.index)):
        img_num +=1 
        img = batch.data[0].asnumpy()[j]
        print(img.shape)
        label = batch.label[0].asnumpy()[j].reshape(7,7,9)
        img4show = np.uint8(img.transpose((1,2,0)))
        bboxdraw(img4show, label) 
print(img_num)        
```
